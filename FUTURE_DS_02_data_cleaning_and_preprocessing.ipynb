```python
#importing libraries and reading the dataset

import pandas as pd
import numpy as np

df = pd.read_csv("Telco_Customer_Churn_Dataset.csv")

print("Initial Shape:", df.shape)
print(df.head())
print()
print(df.info())
```

```python
#Converting Yes/No columns to binary values for easier analysis

yes_no_cols = [ 'Partner', 'Dependents', 'PhoneService', 'PaperlessBilling', 'Churn']
for col in yes_no_cols:
    df[col] = df[col].map({'Yes': 1, 'No': 0})
```

```python
#Feature Engineering
```

```python
df['churn_flag'] = df['Churn']

# Creating groups out of raw values in 'tenure' group for CUSTOMER LIFETIME METRICS
df['tenure_group'] = pd.cut(df['tenure'], bins=[0, 12, 24, 48, 72], labels=['0-1yr', '1-2yr', '2-4yr', '4-6yr'])
```

```python
# Segementing the monthly charges
df['charge_group'] = pd.cut( df['MonthlyCharges'], bins=[0, 50, 100, 150], labels=['Low', 'Medium', 'High'])
```

```python
#Customer value approximation
df['customer_lifetime_value'] = df['MonthlyCharges'] * df['tenure']

df['Contract_Type'] = df['Contract']
```

```python
# Overall churn rate
churn_rate = df['churn_flag'].mean()
print("Churn Rate:", round(churn_rate * 100, 2), "%")

# Average tenure
print("Average Tenure:", round(df['tenure'].mean(), 2))

# Average monthly charges
print("Average Monthly Charges:", round(df['MonthlyCharges'].mean(), 2))
```

```python
# Cohort Analysis Preparation
df['cohort_group'] = df['tenure_group']

# Group for cohort analysis
cohort_data = df.groupby(['cohort_group', 'churn_flag'])['customerID'].count().reset_index()

print(cohort_data.head())

#Retention Rate
df['retained_flag'] = 1 - df['churn_flag']

retention_rate = df['retained_flag'].mean()
print("Retention Rate:", round(retention_rate * 100, 2), "%")
```

```python
print("Final Shape:", df.shape)
print(df.head())
```
